```python

#print("Whats up python!")
# importing calendar module
import calendar
import pandas as pd
from currency_converter import CurrencyConverter
c = CurrencyConverter("http://www.ecb.europa.eu/stats/eurofxref/eurofxref.zip")
# Program make a simple calculator
# This function adds two numbers 
def add(x, y):
   return x + y
# This function subtracts two numbers 
def subtract(x, y):
   return x - y
# This function multiplies two numbers
def multiply(x, y):
   return x * y
# This function divides two numbers
def divide(x, y):
   return x / y
# This function converts pounds to kilograms
def lbstokg(x):
    return x/2.205
# This function converts kilograms to pounds
def kgtolbs(x):
    return x*2.205
# This function converts miles to kilometers
def mitokm(x):
    return x*1.6
# This function converts kilometers to miles 
def kmtomi(x):
    return x/1.6
# This function converts celcius to farenheit 
def ctof(x):
    return (x*9/5)+32
# This function covnerts fareneheit to celcius
def ftoc(x):
    return (x-32)*(5/9)

print("\n")
print("We have two apps for you:")
print("1.Calculator")
print("2.Calendar")
print("3.Weight Converter")
print("4.Distance Converter")
print("5.Temperature Converter")
print("6.Currency Converter")
app = input("Select one app from above\n (1 for Calculator 2 for calender 3 for weight converter 4 for distance converter 5 for temperature converter 6 for currency converter): ")
if (app.isnumeric()==False or (int(app)<0 or int(app)>6)):
        raise Exception("Sorry, you need to choose numbers between 1 and 6.")
if app == '1':
    print("Thanks for choosing calculator, \n please Select one operation from below:")
    print("1.Add")
    print("2.Subtract")
    print("3.Multiply")
    print("4.Divide")
    choice = input("Enter choice(1/2/3/4): ")
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if choice == '1':
        print(num1,"+",num2,"=", add(num1,num2))
    elif choice == '2':
        print(num1,"-",num2,"=", subtract(num1,num2))
    elif choice == '3':
        print(num1,"*",num2,"=", multiply(num1,num2))
    elif choice == '4':
        while True:
            try:
                print(num1,"/",num2,"=", divide(num1,num2))
                break
            except ZeroDivisionError:
                print("Division by 0 is not possible! Please try again!!")
                break
    else:
        print("Invalid input")
elif app == '2':
    # To take month and year input from the user
    yy = int(input("Enter year: "))
    mm = int(input("Enter month: "))

# display the calendar
    print("\n")
    print(calendar.month(yy, mm))
elif app == '3':
    print("Thanks for choosing Weight converter,\n please Select one of the options from below:")
    print("1.Pound")
    print("2.Kilograms")
    choice = input("Enter choice(1/2): ")
    if (choice.isnumeric()==False or (int(choice) < 1 or int(choice) > 2)):
        raise Exception("Sorry you need to enter numbers 1 or 2.")
    elif (choice == "1"):
        print("You have chosen pounds")
        pounds_input=input("Enter the weight in pounds: ")
        if (pounds_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numerical values.")
        conversion = lbstokg(float(pounds_input))
        print('%0.1f weight in pounds is equal to %0.1f weight in kilograms' %(float(pounds_input),conversion))
    elif (choice == "2"):
        print("You have chosen kilograms")
        kilo_input=input("Enter the weight in kilograms: ")
        if (kilo_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numerical values.")
        conversion = kgtolbs(float(kilo_input))
        print('%0.1f weight in kilograms is equal to %0.1f weight in pounds' %(float(kilo_input),conversion))
    else:
        print("Invalid Input")
elif app == '4':
    print("Thanks for choosing Distance converter,\n please Select one of the options from below:")
    print("1.Miles")
    print("2.Kilometers")
    choice = input("Enter choice(1/2): ")
    if (choice.isnumeric==False or(int(choice) < 1 or int(choice) > 2)):
        raise Exception("Sorry you need to enter numbers 1 or 2.")
    elif (choice == "1"):
        print("You have chosen miles")
        miles_input=input("Enter the distance in miles: ")
        if(miles_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numerical values.")
        conversion = mitokm(float(miles_input))
        print('%0.1f distance in miles is equal to %0.1f distance in kilometers' %(float(miles_input),conversion))
    elif (choice == '2'):
        print("You have chosen kilometers")
        kilo_input=input("Enter the distance in kilometers: ")
        if(kilo_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numerical values.")
        conversion = kmtomi(float(kilo_input))
        print('%0.1f distance in kilometers is equal to %0.1f distance in miles' %(float(kilo_input),conversion))
    else :
        print("Invalid Input")
elif app == '5':
    print("Thanks for choosing Temperature converter,\n please Select one of the options from below:")
    print("1.Celcius")
    print("2.Farenheit")
    choice = input("Enter choice(1/2): ")
    if (choice.isnumeric()==False or (int(choice) < 1 or int(choice) > 2)):
        raise Exception("Sorry you need to enter numbers 1 or 2.")
    elif (choice == "1"):
        print("You have chosen celcius")
        celcius_input=input("Enter the temperature: ")
        if(celcius_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numerical values.")
        conversion_expression = ctof(float(celcius_input))
        print('%0.1f degree Celsius is equal to %0.1f degree Fahrenheit' %(float(celcius_input),conversion_expression))
    elif (choice == "2"):
        print("You have chosen farenheit")
        faren_input=float(input("Enter the temperature: "))
        if(faren_input.isnumeric()==False):
            raise Exception("Sorry you need to enter numeric values.")
        conversion_expression = ftoc(float(faren_input))
        print('%0.1f degree Farenheit is equal to %0.1f degree Celcius' %(float(faren_input),conversion_expression))
    else:
        print("Invalid Input")
elif app == '6':
    print("Thanks for choosing Currency converter,\n")
    listu = ['USD', 'JPY', 'BGN', 'CZK', 'DKK', 'GBP', 'HUF', 'PLN',
            'RON', 'SEK', 'CHF', 'ISK', 'NOK', 'HRK', 'RUB', 'TRY', 'AUD',
            'BRL', 'CAD', 'CNY', 'HKD', 'IDR', 'ILS', 'INR', 'KRW', 'MXN',
            'MYR', 'NZD', 'PHP', 'SGD', 'THB', 'ZAR']
    print(listu)
    first = input("Enter the currency you want to enter the value for (look to the reference list above): ")
    second = input("Enter the currency you want it to be exchanged to (look to the reference list above): ")
    if (first in listu and second in listu):
        amount = float(input("Enter the amount: "))
        print("{} {} amounts to {} {}".format(amount,first,c.convert(amount,first,second),second))
    else:
        print("Please enter the currencies in their correct format")
else:
    print("Invalid Input")   
```
